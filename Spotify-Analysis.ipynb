# 2. Further analysis 

```python
df_track.set_index("release_date", inplace = True)
df_track.index= pd.to_datetime(df_track.index)
df_track.head()
```

```python
# So we have songs which were released in 1922
```

```python
df_track["duration"] = df_track["duration_ms"].apply(lambda x: round(x/1000))
```

```python
df_track.drop("duration_ms", inplace=True, axis=1)
df_track.duration.head()
```

# 3. Starting to find insights from data by analyzing data

```python
corr_df = df_track.drop(["key","mode","explicit"], axis=1).corr(method="pearson")
plt.figure(figsize=(20,9))
heatmap= sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=1, vmax=1, center=0, cmap="rocket", linewidths=1, linecolor="Black")
heatmap.set_title("Correlation HeatMap Between Variables")
```

```python
# Conclusion-1:
# There is a relation between energy and loudness.
```

```python
# Conclusion-1:

# Liveliness and violence has least correlation.
```

```python
sample_df = df_track.sample(int(0.004*len(df_track)))
len(sample_df)
```

### The 0.4% is 2346 rows of data 

```python
plt.figure(figsize=(20,12))
sns.regplot(data= sample_df, y= "loudness", x="energy", color = "c").set(title = "Loudness and Energy Correlation")
```

### Song which indicates louder songs are more energetic.

```python
plt.figure(figsize=(20,12))
sns.regplot(data= sample_df, y= "popularity", x="acousticness", color = "b").set(title = "Popularity and Energy Acousticness")
```

### So most acoustic songs are not so popular on spotify.

```python
# create new column of year of releasing date for more insights

df_track["dates"]=df_track.index.get_level_values("release_date")
df_track.dates = pd.to_datetime(df_track.dates)
years=df_track.dates.dt.year
```

```python
# Find Number of Songs by year of release
sns.displot(data = years, discrete = True, aspect=2, height=10, kind="hist").set(title="Number of songs per year")
```

# Conclusion:

### -There is a gradual increase in songs release by every year till 2000.
### -There is a sudden decrease in number of songs in 2000 then again it starts increasing.
### -In 2020 there are most number of songs release.

### See which years have long duration of songs.

```python
# find duration of songs

duration_df = df_track.duration
fig_dims = (21,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = duration_df, ax=ax, errwidth=False).set(title= "Year VS Duration")
plt.xticks(rotation=90)
```
